# FRQ Analysis + MC Reflection (Sprint 9)

## Overview

During Sprint 9, I worked extensively with AP CSA Free Response Questions (FRQs). Initially, these problems appeared difficult due to their length and the density of information in each prompt. However, over time, I recognized that the challenge was not primarily computational complexity, but rather *problem decomposition* and *translation of requirements into structured code*.

At the same time, reflecting on my multiple-choice (MC) performance revealed an important distinction in my skill set: I demonstrate strong conceptual understanding of algorithms and logic, but experience inconsistency when translating that understanding into precise Java syntax and structure. :contentReference[oaicite:0]{index=0}

---

## Evolution of My Approach

At the beginning of FRQ practice, my default strategy was to immediately begin coding. This often resulted in incomplete solutions or logical inconsistencies, especially in multi-step problems.

Through repeated practice, I refined my workflow into a more structured process:

- Careful prompt analysis (multiple readings)
- Identification of required inputs, outputs, and constraints
- Pseudocode development to outline algorithmic steps
- Incremental implementation in Java

This approach significantly improved both correctness and efficiency.

---

## Recognizing Structural Patterns

One of the most impactful realizations was that FRQs follow consistent computational patterns. Most problems can be generalized as:

```input → iterative/conditional processing → output```



This aligns closely with core algorithmic design principles such as iteration, selection, and state tracking. Recognizing these patterns reduced cognitive load and allowed me to focus on implementation details rather than problem interpretation.

---

## Key Conceptual Areas

### ArrayLists and Dynamic Data Structures

FRQs involving `ArrayList` emphasized understanding *mutability* and *iteration behavior*.

**Challenges:**
- Modifying a list during traversal (index shifting issues)
- Distinguishing between element values and indices

**Development:**
- Improved control over loop invariants
- Greater awareness of how structural changes affect iteration

---

### String Manipulation and Parsing

String-based problems required precise index management and careful boundary handling.

**Challenges:**
- Off-by-one errors in substring operations
- Misuse of methods such as `indexOf()`
- Failure to account for edge cases

**Development:**
- More deliberate validation of indices
- Step-by-step tracing of string operations

---

### Object-Oriented Design

Class-based FRQs reinforced foundational OOP principles.

**Key areas:**
- Constructor design
- Encapsulation through private fields
- State management across method calls

This aligned with broader object-oriented concepts such as abstraction and encapsulation, which are also emphasized in software engineering practices. :contentReference[oaicite:1]{index=1}

---

### Multi-Step Algorithmic Reasoning

More complex FRQs required chaining multiple operations together.

These problems often involved:
- Nested iteration
- Conditional state tracking
- Sequential transformations of data

The most effective strategy was decomposing the problem into smaller, independent steps before integrating them into a complete solution.

## Final Reflection

Initially, FRQs felt unstructured and difficult to approach. Through practice, they have become more predictable and systematic.

The most significant shift in my thinking is:

> FRQs are not isolated problems; they are structured applications of core algorithmic and object-oriented principles.

While I still need to improve execution and syntax fluency, I now approach these problems with a clear, repeatable strategy grounded in problem decomposition and algorithm design.
